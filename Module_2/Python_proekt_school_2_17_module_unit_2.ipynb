## Функция get_info_about_column 

```python
# функция get_info_about_column - возвращает информацию по колонке
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)
#          plot - 1 - строить график; 0 - не строить (из-за ошибок) - (int)
#          hist - 0 - строить обычный график; 1 - гистограммой - (int)
# Возвращает
#           - выводит информацию о заполненых полях (class 'pandas.core.frame.DataFrame)
#           - выводит данные по распределению значений
#           - количество уникальных значений в столбце (int)
#           - расшифровку этих значений (pandas.core.frame.DataFrame)
#           - построенный график уникальных значений (list)


def get_info_about_column(table, column_name, plot=1, hist=0):

    print('Данные по заполненности полей:')
    print('')
    display(table.loc[:, [column_name]].info())
    print('')
    print('Данные о распредлении значений:')
    print('')
    display(table[column_name].describe())
    print('')
    print('Уникальных значений в колонке "' + column_name + '": ',
          table[column_name].nunique())
    display(pd.DataFrame(table[column_name].value_counts()))
    if plot == 1:
        if hist == 1:
            table[column_name].hist()
        else:
            plt.plot(table[column_name])
    if table[column_name].isnull().any() == True:
        print('Количество не заполненных значений: ',
              table[column_name].isnull().sum())
    else:
        print('Нет не заполненных значений в колонке. ')
```

##  Функция get_column_range

```python
# функция get_column_range - возвращает выбросы значений по колонке
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)
# Возвращает значение выбросов (string)


def get_column_range(table, column_name):
    median = table[column_name].median()
    perc25 = table[column_name].quantile(0.25)
    perc75 = table[column_name].quantile(0.75)
    IQR = perc75 - perc25
    print('Разброс значений в колонке "' + column_name + '": ')
    print('')
    print(
        '25-й перцентиль: {},'.format(perc25),
        '75-й перцентиль: {},'.format(perc75), "IQR: {}, ".format(IQR),
        "Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5 * IQR,
                                               l=perc75 + 1.5 * IQR))
```

## Функция fill_null_value_in_column

```python
# функция fill_null_value_in_column - заполняет пустые значения в колонке на выбранное значение
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)
#          null_value - представление пустого значения -(string)
#          fill_value - представление значения для заполнения -(string)
# Возвращает - ничего не возвращает


def fill_null_value_in_column(table,
                              column_name,
                              fill_value=None):
    if type(0.0) == type(fill_value):
        table[column_name] = table[column_name].apply(
            lambda x: fill_value if pd.isnull(x) else fill_value if x == 'nan' else x)
    else:
        table[column_name] = table[column_name].apply(
            lambda x: fill_value if pd.isnull(
                x) else fill_value if x == 'nan' else x
            if '(' not in x else x[:x.find('(')].strip())
```

## функция replace_value_in_column

```python
# функция replace_value_in_column - заменяет значение в колонке на выбранное значение
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)
#          value - значение которое будет заменяться
#          replace_value - значение на котрое будет произведена замена
# Возвращает - ничего не возвращает


def replace_value_in_column(table,
                            column_name,
                            value,
                            replace_value):
    table[column_name] = table[column_name].apply(
        lambda x: replace_value if x == value else x)
```

## Функция null_value_in_column

```python
# функция null_value_in_column - выбирает все строки,
# в которых в столбце есть не заполненные значения null
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)


def null_value_in_column(table, column_name):
    bool_series_isnull = pd.isnull(table[column_name])
    table[bool_series_isnull]
```

## Функция notnull_value_in_column

```python
# функция notnull_value_in_column - выбирает все строки,
# в которых в столбце нет не заполненных значений null
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_name - имя столбца (string)


def notnull_value_in_column(table, column_name):
    bool_series_not_null = pd.notnull(table[column_name])
    table[bool_series_not_null]
```

## Функция get_boxplot

```python
# функция get_boxplot - строит график для номинативных колонок
# column_x, column_y таблицы table
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_x - имя столбца (string) - номинальная переменная
#          column_y - имя столбца (string) - значимая перменная


def get_boxplot(table, column_x, column_y):
    fig, ax = plt.subplots(figsize=(14, 4))
    sns.boxplot(x=column_x, y=column_y,
                data=table,
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column_x)
    plt.show()
```

## Функция get_stat_dif

```python
# функция get_stat_dif - строит график для колонок
# column_x, column_y таблицы table
# Параметры
#          table - таблица (pandas.core.frame.DataFrame),
#                  в которой находится колонка
#          column_x - имя столбца (string) - номинальная переменная
#          column_y - имя столбца (string) - значимая перменная


def get_stat_dif(table, column_x, column_y):
    cols = table.loc[:, column_x].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(table.loc[table.loc[:, column_x] == comb[0], column_y],
                     table.loc[table.loc[:, column_x] == comb[1], column_y]).pvalue \
                <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column_x)
            break
```

## Функция get_corr_value

```python
# функция get_corr_value - выводит значения корреляций в удобной форме
#
# Параметры
#    corr_tab - pandas.core.frame.DataFrame - полученная в результате corr()


def get_corr_value(corr_tab):
    dic_corr = {}
    pozitve_corr = []
    pozitive_non_corr = []
    negative_corr = []
    negative_non_corr = []

    for r in range(len(corr_tab.columns)):
        for c in range(len(corr_tab.iloc[r].index)):
            value = corr_tab.iloc[r][corr_tab.iloc[r].index[c]]
            name_row = corr_tab.iloc[r].index[r]
            name_column = corr_tab.iloc[r].index[c]
            if value == 1.0:
                continue
            if value not in dic_corr:
                dic_corr[value] = {}
                dic_corr[value]['row'] = name_row
                dic_corr[value]['column'] = name_column
                if value >= 0.1:
                    pozitve_corr.append(value)
                elif value > 0 and value < 0.1:
                    pozitive_non_corr.append(value)
                elif value <= -0.1:
                    negative_corr.append(value)
                elif value < 0 and value > -0.1:
                    negative_non_corr.append(value)

    pozitve_corr.sort(reverse=True)
    pozitive_non_corr.sort(reverse=True)
    negative_corr.sort(reverse=False)
    negative_non_corr.sort(reverse=False)

    print('ПОЛОЖИТЕЛЬНУЮ КОРРЕЛЯЦИЮ ИМЕЮТ:')
    print('')
    for i in pozitve_corr:
        print(i, ':', dic_corr[i]['row'], '-', dic_corr[i]['column'])

    print('')
    print('СЛАБУЮ ПОЛОЖИТЕЛЬНУЮ КОРРЕЛЯЦИЮ ИМЕЮТ:')
    print('')
    for i in pozitive_non_corr:
        print(i, ':', dic_corr[i]['row'], '-', dic_corr[i]['column'])

    print('')
    print('ОТРИЦАТЕЛЬНУЮ КОРРЕЛЯЦИЮ ИМЕЮТ:')
    print('')
    for i in negative_corr:
        print(i, ':', dic_corr[i]['row'], '-', dic_corr[i]['column'])

    print('')
    print('СЛАБУЮ ОТРИЦАТЕЛЬНУЮ КОРРЕЛЯЦИЮ ИМЕЮТ:')
    print('')
    for i in negative_non_corr:
        print(i, ':', dic_corr[i]['row'], '-', dic_corr[i]['column'])
```
